# Нечеткое сравнение + cities

```python
from fuzzywuzzy import process
from fuzzywuzzy import fuzz
import csv
import os
```

```python
TEMP_DIR = 'temp'
# в кодировке Windows-1251
file_cities_csv = os.path.join(TEMP_DIR, "cities.csv")
# в кодировке utf-8
file_cities_utf8_csv = os.path.join(TEMP_DIR, "cities_utf8.csv")
```

```python
with open(file_cities_csv) as f:
    reader = csv.DictReader(f, delimiter=';')
    for line in reader:
        print(line['Город'])
```

```python
search_text = 'Воронеж'

with open(file_cities_csv) as f:
    reader = csv.DictReader(f, delimiter=';')
    for line in reader:
        print(f"{line['Город']} - {fuzz.WRatio(search_text, line['Город'])}")
```
