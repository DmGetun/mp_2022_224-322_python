## Простые графики

```python
fig, ax = plt.subplots()
ax.plot([0,1,2,3,4,5], [2,3,10,5,7,1])
ax.plot([0,1,2,3,4,5], [0,1,3,2,6,5], linewidth=5)
ax.scatter([1,5],[9,3], color='red')
plt.show()
```

```python
x = [0,1,2,3,4,5]
y1 = [2,3,10,5,7,1]
y2 = [0,1,3,2,6,5]

fig, ax = plt.subplots()
ax.plot(x, y1)
ax.plot(x, y2, linewidth=5)
ax.scatter([1,5],[9,3], color='red',marker='*')
plt.show()
```

```python
x = [i for i in range(30)]
y = [i**6 for i in range(30)]

plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

fig, ax = plt.subplots()
ax.set_title('График Sin(x) и Cos(x)')
ax.plot(x,y_sin, label='sin')
ax.plot(x,y_cos,label='cos')
ax.axhline(0, color='green')

plt.legend() # показать легенду
ax.grid() # показать сетку
plt.show()
```

## Разместить несколько графиков

```python
fig = plt.figure()
fig.set_facecolor('lightgreen')

# строка, колонка, позиция
ax_1 = fig.add_subplot(2, 1, 1)
ax_1.plot(x,y_sin, label='sin')
ax_2 = fig.add_subplot(2, 2, 3)
ax_3 = fig.add_subplot(2, 2, 4)

ax_1.set(title='График 1',xticks=[], yticks=[])
ax_2.set(title='График 2', xticks=[], yticks=[])
ax_3.set(title='График 3', xticks=[], yticks=[])
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0,0].set(title='График [0,0]')

for ax in axes.flat:
    ax.set(xticks=[],yticks=[])

plt.show()
```

```python
labels = ['Еда', 'Кафе', 'Перекусы', 'Напитки']
values = [40, 10, 20, 30]

fig, ax = plt.subplots()
fig.set_facecolor('lightgreen')
ax.set_title('Мои расходы')
ax.pie(values,labels=labels, wedgeprops={'linewidth': 5, 'edgecolor': 'white'})
plt.show()
```
