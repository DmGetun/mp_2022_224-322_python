## Группировка данных

```python
df.Nationality.unique()
```

```python
df[df.Nationality == 'Argentina'].Age
```

```python
group1 = df[df.Nationality == 'Argentina'].Age
group2 = df[df.Nationality == 'Brazil'].Age
group3 = df[df.Nationality == 'Portugal'].Age

df_new = pd.DataFrame(
    [group1.mean(), group2.mean(), group3.mean()],
    columns = ['Age'],
    index=['Argentina', 'Brazil','Portugal']
)
df_new
```

```python
df_new = pd.DataFrame(
    {'Age': [group1.mean(), group2.mean(), group3.mean()],
     'Min': [group1.min(), group2.min(), group3.min()],
     'Max': [group1.max(), group2.max(), group3.max()]},
    index=['Argentina', 'Brazil','Portugal']
)
df_new
```

## Сводные таблицы

```python
df[df['Club'].isin(['FC Barcelona', 'Juventus', 'Manchester United', 'Al Batin', 'Manchester City'])]
```

```python
# margins=True - столбец ALL
# fill_value=0 - заменить NaN на 0
pivot = df[df['Club'].isin(['FC Barcelona', 'Juventus', 'Manchester United', 'Al Batin', 'Manchester City'])].pivot_table(
    values=['Wage'], 
    index=['Nationality'], 
    columns=['Club'],
    aggfunc='sum',
    margins=True,
    fill_value=0
)
pivot
```
